* This data used in this program comes from http://hospitalcompare.hhs.gov, and contains information about the quality of care of over 4000 Medicare-certified hospitals in the US. Quality of care of each hospital will be assessed here by  the outcome "30 Day death rate from heart attack/ pneumonia/ heart failure".

* This function takes in 2 arguements, an outcome measure (heart attack, heart failure or pneumonia) and a rank. The function will return a pandas dataframe which contains, for EACH state, the hospital at the specified rank for the specified outcome.

* Eg. rankall("pneumonia", 5) will return a dataframe that lists for each state the hospital that is ranked 5 in that state for number of deaths due to pneumonia. Please note that the higher the rank, the lower is the mortality.

```python
import pandas as pd

def rankall(outcome, num = "best"):
    
    df = pd.read_csv("C:/Users/Peeka/Desktop/Codes/outcome-of-care-measures.csv", na_values = "Not Available")
        
    
    uniques = df["State"].unique().tolist()
    hospitaldictionary = {}
    statedictionary = {}
    
    if outcome == "heart attack":
        outcomerow = df.columns[10]
    elif outcome == "heart failure":
        outcomerow = df.columns[16]
    elif outcome == "pneumonia":
        outcomerow = df.columns[22]
    else:
        return "invalid outcome"
    
    for i in uniques:
        subset = df.loc[df.State == i, ["State", "Hospital Name", outcomerow]]
        subset = subset.sort_values([outcomerow, "Hospital Name"], ascending = [True, True])
        subset = subset.dropna()
        
        if num == "best":
            wanted = subset.iloc[0, 1]
        elif num == "worst":
            wanted = subset.iloc[subset.shape[0] - 1, 1]
        else:
            if num > subset.shape[0]:
                wanted = "NA"
            else:
                wanted = subset.iloc[num - 1, 1]

            
        hospitaldictionary[i] = wanted
        statedictionary[i] = i
    
    hospital = pd.Series(hospitaldictionary)
    state = pd.Series(statedictionary)
    answer = pd.DataFrame({"Hospital": hospital, "State": state})
    
    return answer

rankall("heart attack", 20)
```
